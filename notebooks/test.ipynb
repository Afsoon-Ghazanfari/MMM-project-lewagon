```python
import matplotlib.pyplot as plt
import mplcursors
import ipywidgets as widgets
from IPython.display import display

# Assuming fb_data_monthly is your DataFrame containing the required columns

# Create a slider for controlling the x-axis range
slider = widgets.IntRangeSlider(
    value=[0, len(fb_data_monthly)-1],
    min=0,
    max=len(fb_data_monthly)-1,
    step=1,
    description='Date Range:',
    continuous_update=False  # Update only when slider released
)

# Create the interactive figure
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the 'costs' and 'clicks' columns on the primary y-axis
lines = []
lines.append(ax.plot(fb_data_monthly.index, fb_data_monthly['fb_costs'], marker='o', color='blue', label='Costs')[0])
lines.append(ax.plot(fb_data_monthly.index, fb_data_monthly['fb_clicks'], marker='o', color='green', label='Clicks')[0])

# Set plot title and labels
ax.set_title('Monthly Clicks and Costs')
ax.set_xlabel('Date')
ax.set_ylabel('Counts')
ax.legend()

# Create a twin y-axis for the impressions data
ax2 = ax.twinx()
lines.append(ax2.plot(fb_data_monthly.index, fb_data_monthly['fb_impressions'], marker='o', color='orange', label='Impressions')[0])
ax2.set_ylabel('Impressions')

# Attach mplcursors to the figure
mplcursors.cursor(lines)

# Function to update the plot based on slider value
def update_plot(change):
    start_idx, end_idx = slider.value
    subset_data = fb_data_monthly.iloc[start_idx:end_idx+1]
    
    ax.clear()  # Clear the previous plot

    # Plot the 'costs' and 'clicks' columns on the primary y-axis
    lines = []
    lines.append(ax.plot(subset_data.index, subset_data['fb_costs'], marker='o', color='blue', label='Costs')[0])
    lines.append(ax.plot(subset_data.index, subset_data['fb_clicks'], marker='o', color='green', label='Clicks')[0])

    # Set plot title and labels
    ax.set_title('Monthly Clicks and Costs')
    ax.set_xlabel('Date')
    ax.set_ylabel('Counts')
    ax.legend()

    # Create a twin y-axis for the impressions data
    ax2 = ax.twinx()
    lines.append(ax2.plot(subset_data.index, subset_data['fb_impressions'], marker='o', color='orange', label='Impressions')[0])
    ax2.set_ylabel('Impressions')

    # Attach mplcursors to the updated plot
    mplcursors.cursor(lines)

    plt.xticks(rotation=45)
    
    plt.tight_layout()

# Attach the update function to the slider's value change event
slider.observe(update_plot, names='value')

# Initial plot
update_plot(None)

# Display the slider and the plot
display(slider)
display(fig)
```
